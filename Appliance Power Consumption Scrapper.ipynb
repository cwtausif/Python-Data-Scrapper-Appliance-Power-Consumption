<h1>Appliance Power Consumption Scrapper</h1>

```python
url = "https://www.daftlogic.com/information-appliance-power-consumption.htm"
url
```

```python
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from urllib.request import Request, urlopen

req = Request(
    url=url, 
    headers={'User-Agent': 'Mozilla/5.0'}
)
page = urlopen(req).read()
page
title = str(page).split('<title>')[1].split('</title>')[0]
title
```

```python
soup = BeautifulSoup(page, 'html.parser')
```

```python
table = soup.find('table', id="tblApp")
table
```

```python
df = pd.read_html(str(table))[0]
```

```python
df.head()
```

```python
df.to_csv("appliances.csv")
```
